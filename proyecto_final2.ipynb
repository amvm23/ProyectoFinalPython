```python
def regresion (x, y, i, j_error, rl, mi, bi):
    
    key = []
    error = []
    resultado = []

    
    for iteracion in range(i):
        
        p = np.array([mi,bi])

        unos = np.ones(shape=(x.shape))
        a=np.reshape(x,(x.shape[0],1))
        b=np.reshape(unos,(unos.shape[0],1))

        xv = np.hstack((a,b))

        yhat = np.dot(xv,p)


        e = 0.5*np.average(pow(y-yhat,2))
        error.append(e)
        key.append(iteracion)

        gradiente_m = np.average((yhat-y)*x1)
        gradiente_b = np.average((yhat-y))

        mi = mi - lr*gradiente_m
        bi = bi - lr*gradiente_b

    return np.array(key), np.array(yhat), np.array(error), mi, bi, e


   
```

```python
def grafica_error(key, error):
    kg = np.array(key)
    eg = np.array(error)

    plt.plot(kg,eg)

    plt.xlabel('iteracion')
    plt.ylabel('error')
    plt.title('Grafica de Error')

    plt.show()
```

```python
def evolucion(x, y, yhat):
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1) #creamos un canvas con  1 fila, 1 columna y devolvemos una referencia a la primera grafica

    ax.grid()
    ax.scatter(x,y,color="b")
    ax.plot(x,yhat,label="yhat")

    ax.legend()
    ax.set_xlabel('datos X')
    ax.set_ylabel('datos Y')
    ax.set_title('Grafica de y = mx + b')

    plt.show()
```

```python
#Variable independiente OverallQual

iteracion, resultado_yhat, resultado_error, m, b, e = regresion(x1, y, 1000, 10, 0.1, 0.1, 0.1)  
print("parametros m y b: ", m, ", ", b)
print("\nerror: ", e)
grafica_error(iteracion, resultado_error)
evolucion(x1, y, yhat)
```

```python
#Variable independiente 1stFlrSF
iteracion, resultado_yhat, resultado_error, m, b, e = regresion(x3, y, 100, 10, 0.1, 0.1, 0.001)
print("parametros m y b: ", m, ", ", b)
print("\nerror: ", e)
grafica_error(iteracion, resultado_error)
evolucion(x3, y, yhat)
```
